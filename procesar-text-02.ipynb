## Processament del llenguatge natural. Exemple 02

Anem a veure la llibreria `spacy`, que és una altra eina molt potent per al processament del llenguatge natural en Python. Per algunes coses és més completa que `nltk` i és més ràpida, però també és més complexa d'utilitzar. En aquest exemple, veurem com utilitzar `spacy` per analitzar un text i extreure informació útil.

```python
# Anem a importar les llibreries necessàries per a treballar amb spacy i fer l'anàlisi de text
import spacy
# Carreguem el model de llenguatge per a castellà
nlp = spacy.load("es_core_news_sm")
```

Anem a repetir l'exemple anterior de El Quijote, que hem fet amb NLTK, però esta vegada amb Spacy. Així comparem quines coses funcionen igual, quines millor i si alguna funciona pitjor.

```python
import requests

url = "https://babel.upm.es/~angel/teaching/pps/quijote.txt"
resp = requests.get(url, verify=False)  # ← Ignora SSL
resp.raise_for_status()

texto = resp.text
print(texto[:500])  # Primeros 500 chars
print("\nLongitud:", len(texto))
```

```python
# Les expressions regulars per a tokenizar amb spaCy no són necessàries, ja que spaCy té
#  un tokenitzador integrat molt potent que treballa de manera eficient i precisa. 
# Simplement passem el text al model de llenguatge carregat i este s'encarrega de dividir
#  el text en tokens, tenint en compte les regles gramaticals i les particularitats del llenguatge.
#  Això ens permet obtenir una tokenització més natural i adequada al context del text, 
# sense haver de preocupar-nos per les complexitats de les expressions regulars.

# Primer processem el text

# per a que no ens done problemes, ajustem la longitud màxima del text que spacy pot processar
nlp.max_length = len(texto) + 1000 # un marge per a evitar problemes

doc = nlp(texto)

# Amb això ja podem demanar oracions, tokens, etc. sense fer res més

# Ara veiem les oracions

    # mostrem les primeres 5 oracions
# cada oració en una línia
for sentence in list(doc.sents)[:5]:
    print(sentence)
    # un separador, per veure on acaba cada oració
    print("-" * 50)

# Mostrem les 50 primeres paraules 
print(f"\n50 primers tokens:")
for token in doc[:50]: 
    print(token.text)


```

```python
# Hem vist que de manera automàtica spacy tokenitza també els signes de puntuació i caràcters especials.
# Una forma de evitar-ho seria processar el text sense els signes de puntuació, però això ens impedirà
# detectar oracions.

# Per tant, anem a eliminar els tokens que siguen signes de puntuació, espais en blanc o altres caràcters especials.
# Fixeu-vos que tampoc cal eliminar els stop words, perquè spacy ja sap en quin idioma està treballant
# i les detecta automàticament.
filtered_tokens = [token for token in doc if not (token.is_punct or token.is_space or token.is_stop)]

# Mostrem els primers 50 tokens després de la neteja que hem fet.

print(f"\n50 primers tokens després de la neteja:")
for token in filtered_tokens[:50]: print(token)
```

```python
# De moment el resultat que ens torna spaCy després de tokenitzar i eliminar caràcters especials
# i stop words és el mateix que hem obtés en NLTK.

# Anem a veure si l'etiquetatge el fa millor.

# Per cert, també el fa automàticament, així que només cal mostrar el resultat
# Utilitzem filtered_tokens que ja té els tokens netejats

for token in filtered_tokens[:50]:  # Només els 50 primers
    print(f"{token.text:15} | {token.pos_:7} | {token.tag_:9} | {spacy.explain(token.tag_)}")
```

No és perfecte però molt millor que amb NLTK.

```python
# Ara anem a fer stemming i lemmatització al mateix temps.

# SpaCy no té un stemmer integrat, així que anem a fer una barreja de SpaCy i NLTK.
# Utilitzarem SnowBall que abans hem vist que funciona bé en castellà.

from nltk.stem.snowball import SnowballStemmer

stemmer = SnowballStemmer('spanish')

for token in filtered_tokens[:50]: # Només els 50 primers
    if token.is_alpha:
        lemma = token.lemma_
        stem = stemmer.stem(token.text.lower())
        # El lemma no cal calcular-lo perquè spaCy ja ho fa automàticament
        print(f"{token.text:12} | Lema: {lemma:10} | Stem: {stem}")

```

Podem veure que ho fa millor que NLTK, però no és perfecte del tot tampoc. Per exemple, el lema de "habiendo" és "haber", el de "mandaron" és mandar, i el de "doy" és "dar", però el de "certifico" hauria de ser "certificar" i no l'ha trobat.

Ara anem a fer una anàlisi exploratòria tal com hem fet en l'exemple anterior amb NLTK. De fet ací també utilitzarem NLTK per a algunes coses. 

```python
import spacy
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures
from nltk.probability import FreqDist
from collections import Counter

print("Análisis exploratorio spaCy + NLTK")
print("=" * 60)

# 1. FreqDist (top tokens/lemas)
tokens_lemas = [token.lemma_ for token in filtered_tokens]
fdist = FreqDist(tokens_lemas)
print("Top 10 lemas:", fdist.most_common(10))

# 2. Bigramas (igual que NLTK)
bigram_measures = BigramAssocMeasures()
finder = BigramCollocationFinder.from_words(tokens_lemas)
finder.apply_freq_filter(2)  # Solo bigramas ≥2 veces

print("\nTop 5 bigramas (raw_freq):")
print(finder.nbest(bigram_measures.raw_freq, 5))

print("\nTop 5 bigramas (likelihood_ratio):")
print(finder.nbest(bigram_measures.likelihood_ratio, 5))

# 3. Longitud media
print(f"\nLongitud mitjana lemas: {round(sum(len(w) for w in filtered_tokens) / len(filtered_tokens), 2):.2f}")

# 4. Bonus: Top POS
pos_dist = Counter(token.pos_ for token in doc if token.is_alpha)
print("\nTop POS:", pos_dist.most_common(5))
```
